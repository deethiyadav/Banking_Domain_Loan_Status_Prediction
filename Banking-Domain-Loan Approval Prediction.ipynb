# Data Preparation

## Identify the Target Variable and Splitting the Data into Train and Test

```python
# Identifying the inputs (X) and output (y)
y = df['loan_status']   #Target Variable

X = df[['person_age', 'person_income', 'person_home_ownership',
       'person_emp_length', 'loan_intent', 'loan_grade', 'loan_amnt',
       'loan_int_rate','loan_percent_income',
       'cb_person_default_on_file', 'cb_person_cred_hist_length']]
```

```python
y.head()
```

```python
X.head()
```

```python
# split into train and test

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.75, random_state=100)
```

```python
print(X_train.shape, y_train.shape)

print(X_test.shape, y_test.shape)
```

```python
#Separating Categorical and Numerical Columns
X_train.head()
```

```python
X_train.dtypes
```

```python
X_train_cat = X_train.select_dtypes(include=['object'])

X_train_cat.head()
```

```python
X_train_num = X_train.select_dtypes(include=['int64', 'float64'])

X_train_num.head()
```

```python
# scaling the numerical features
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_num_rescaled = pd.DataFrame(scaler.fit_transform(X_train_num), 
                                    columns = X_train_num.columns, 
                                    index = X_train_num.index)

X_train_num_rescaled.head()
```

```python
X_train_num.describe()
```

```python
import numpy as np
```

```python
print("Number of Numerical Features:", scaler.n_features_in_)
print("Mean of each column:", scaler.mean_)
print("Std of each column:", np.sqrt(scaler.var_))
```

```python
X_train_cat['person_home_ownership'].value_counts(normalize=True)
```

```python
X_train_cat['loan_intent'].value_counts(normalize=True)
```

```python
X_train_cat['loan_grade'].value_counts(normalize=True)
```

```python
X_train_cat['cb_person_default_on_file'].value_counts(normalize=True)
```

```python

from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(drop='first', sparse=False)

# column names are (annoyingly) lost after OneHotEncoding
# (i.e. the dataframe is converted to a numpy ndarray)

X_train_cat_ohe = pd.DataFrame(encoder.fit_transform(X_train_cat), 
                               columns=encoder.get_feature_names_out(X_train_cat.columns), 
                               index = X_train_cat.index)

X_train_cat_ohe.head()
```

```python
X_train_transformed = pd.concat([X_train_num_rescaled, X_train_cat_ohe], axis=1)

X_train_transformed.head()
```

```python
#Preparing test data
X_test.head()
```

```python
X_test.info()
```

```python
X_test_cat = X_test.select_dtypes(include=['object'])

X_test_cat.head()
```

```python
X_test_num = X_test.select_dtypes(include=['int64', 'float64'])

X_test_num.head()
```

```python
X_test_num_rescaled = pd.DataFrame(scaler.transform(X_test_num), 
                                   columns = X_test_num.columns, 
                                   index = X_test_num.index)

X_test_num_rescaled.head()
```

```python
X_test_cat_ohe = pd.DataFrame(encoder.transform(X_test_cat), 
                               columns=encoder.get_feature_names_out(X_test_cat.columns), 
                               index = X_test_cat.index)

X_test_cat_ohe.head()
```

```python
X_test_transformed = pd.concat([X_test_num_rescaled, X_test_cat_ohe], axis=1)

X_test_transformed.head()
```

# Modelling

```python
models = pd.DataFrame(columns=['algorithm','metric','train','test'])
```

```python
from sklearn import metrics
```

```python
metric_funcs = {
    "accuracy": metrics.accuracy_score,
    "precision": metrics.precision_score,
    "recall": metrics.recall_score,
    "f1_score": metrics.f1_score,
    "roc_auc": metrics.roc_auc_score
}
```

# Logistic Regression

```python
%%time
from sklearn.linear_model import LogisticRegression     # x_train_transformed
Lr_model = LogisticRegression()
Lr_model.fit(X_train_transformed, y_train)
```

```python
y_train_pred = Lr_model.predict(X_train_transformed)
y_test_pred = Lr_model.predict(X_test_transformed)
```

```python
for name,func in metric_funcs.items():
    models = models.append({
        "algorithm": "Logistic Regression",
        "metric": name,
        "train": func(y_train,y_train_pred),
        "test": func(y_test,y_test_pred)
    }, ignore_index=True)
```

# KNN Classifier

```python
%%time
from sklearn.neighbors import KNeighborsClassifier     # x_train_transformed
Knn = KNeighborsClassifier()
Knn.fit(X_train_transformed, y_train)
```

```python
y_train_pred = Knn.predict(X_train_transformed)
y_test_pred = Knn.predict(X_test_transformed)
```

```python
for name,func in metric_funcs.items():
    models = models.append({
        "algorithm": "KNN",
        "metric": name,
        "train": func(y_train,y_train_pred),
        "test": func(y_test,y_test_pred)
    }, ignore_index=True)
```

```python
models.pivot_table(index = "algorithm", columns = "metric",values="test")
```

```python
for m in metric_funcs:
    Data = models.loc[models.metric==m]
    Data = Data.set_index("algorithm").drop("metric",axis=1).sort_values(by="test")
    
    Data.plot(kind="bar",colormap="PiYG")
    plt.grid(linestyle="--")
    plt.ylabel(m)
    plt.title(m.upper())
    plt.show()
```
